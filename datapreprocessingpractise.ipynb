```python
dt=dataset.copy()
```

```python
cols=["Gender","Area","Loan_Status"]
```

```python
dt[cols]=dt[cols].fillna(dt.mode().iloc[0])  # filling nan values with mode iloc is given the one or mode values can be given in a coloum so we consider the first mode returned
```

```python
dt.isnull().sum(axis=0)
```

```python
cols2=["ApplicantIncome","CoapplicantIncome","LoanAmount"]
```

```python
dt[cols2]=dt[cols2].fillna(dt.mean())   # fiiling the nan values with mean
```

```python
dt.isnull().sum(axis=0)
```

```python
dt[cols]=dt[cols].astype('category')
dt.dtypes                     # we convert object to category as objects may contain strings also so astype converts it to category.
```
